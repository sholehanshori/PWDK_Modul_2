```python
import pandas as pd
import numpy as np

df = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Andi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Andi', 'fisika': 92, 'biologi': 82},
    {'nama': 'Budi', 'fisika': 90, 'biologi': 95},
    {'nama': 'Budi', 'fisika': 80, 'biologi': 98},
    {'nama': 'Budi', 'fisika': 88, 'biologi': 90}
])
df
```

```python
# Mengambil data menggunakan grup (kesamaan)
dfGroup = df.groupby('nama')
dfGroup.get_group('Andi')
```

```python
# Untuk mengetahui group yg ada
dfGroup.groups.keys()
```

```python
# Untuk mendapatkan nilai 'max' dari setiap grup begitupula untuk 'min' ataupun rata-rata 'mean'
dfGroup.max()
```

```python
# Mendapatkan nilai max pada mata pelajaran tertentu di grup tertentu
dfGroup.get_group('Andi').max()['fisika']
```

```python
# Sama seperti sebelumnya, beda cara
df[df['nama'] == 'Andi']['fisika'].max()
```

```python
# Concat, menggabungkan dengan stacking
df1 = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Budi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Caca', 'fisika': 92, 'biologi': 82}
])

df2 = pd.DataFrame([
    {'nama': 'Didi', 'fisika': 90, 'biologi': 95},
    {'nama': 'Euis', 'fisika': 80, 'biologi': 98},
    {'nama': 'Fafa', 'fisika': 88, 'biologi': 90}
])

dfX = pd.concat([df1, df2], ignore_index= True)
# Memberi ignore_index agar tidak memiliki index yg sama
dfX
```

```python
dfX.loc[0]
```

```python
# Apabila propertinya berbeda antara df3 dan df4

df3 = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Budi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Caca', 'fisika': 92, 'biologi': 82}
])

df4 = pd.DataFrame([
    {'nama': 'Didi', 'math': 90, 'biologi': 95},
    {'nama': 'Euis', 'math': 80, 'biologi': 98},
    {'nama': 'Fafa', 'math': 88, 'biologi': 90}
])

dfY = pd.concat([df3, df4], ignore_index= True, sort=False)
# Memberi ignore_index agar tidak memiliki index yg sama
# Sort=False karena properti ada yg berbeda
dfY
```

```python
# Apabila properti 'nama' berisi sama tetapi properti yg lain berbeda
df5 = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Budi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Caca', 'fisika': 92, 'biologi': 82}
])

df6 = pd.DataFrame([
    {'nama': 'Andi', 'math': 90, 'biologi': 95},
    {'nama': 'Budi', 'math': 80, 'biologi': 98},
    {'nama': 'Caca', 'math': 88, 'biologi': 90}
])

dfZ = pd.concat([df5, df6], ignore_index= True, sort=False)
# Memberi ignore_index agar tidak memiliki index yg sama
# Sort=False karena properti ada yg berbeda
dfZ
```

```python
# Hasil sebelumnya terjadi repetisi, sehingga apabila diberikan 'axis=1'
dfZ = pd.concat([df5, df6], axis=1, ignore_index= True, sort=False)
dfZ
# Hasil menjadi lebih compact, tetapi kurang informatif
```

```python
# Merge = join

df7 = pd.DataFrame([
    {'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'nama': 'Budi', 'fisika': 95, 'biologi': 80},
    {'nama': 'Caca', 'fisika': 92, 'biologi': 82}
])

df8 = pd.DataFrame([
    {'nama': 'Budi', 'math': 90, 'kimia': 95},
    {'nama': 'Andi', 'math': 80, 'kimia': 98},
    {'nama': 'Caca', 'math': 88, 'kimia': 90},
    {'nama': 'Deni', 'math': 88, 'kimia': 90}
])

dfA = pd.merge(df7, df8)
dfA
```

```python
# 'Outer' untuk menampilkan semua data walaupun tidak ada
dfB = pd.merge(df7, df8, on='nama', how='outer')
dfB
# Penggunaan "on='nama'" digunakan agar lock 'properti' yg ingin dipakai patokan
```

```python
# Pivot = Menggabungkan properti data berdasarkan 1 properti yg berulang
df = pd.DataFrame([
    {'test':1, 'nama': 'Andi', 'fisika': 90, 'biologi': 78},
    {'test':2, 'nama': 'Andi', 'fisika': 95, 'biologi': 80},
    {'test':3, 'nama': 'Andi', 'fisika': 92, 'biologi': 82},
    {'test':1, 'nama': 'Budi', 'fisika': 80, 'biologi': 88},
    {'test':2, 'nama': 'Budi', 'fisika': 85, 'biologi': 89},
    {'test':3, 'nama': 'Budi', 'fisika': 72, 'biologi': 72}
])
df = df.pivot(index='test', columns='nama')
df
```

```python
# Memanggil salah satu properti
df['fisika']['Andi'].loc[1]
```

```python
# Membuat tabel secara manual

dfQ = pd.DataFrame({
    'mapel': ['biologi', 'biologi', 'biologi', 'fisika', 'fisika', 'fisika'],
    'test' : [1,2,3, 1,2,3],
    'Andi' : [78,80,82,90,95,92],
    'Budi' : [88,89,72,80,85,72]
})
df01 = dfQ.set_index(['test', 'mapel'])
df01
```

```python
# Menggunakan 'swaplevel' untuk menukar posisi atas/bawah dari header
df02 = df01.unstack().swaplevel(0,1,1).sort_index(1)
df02
```

```python
# Menukar posisi index
df03 = dfQ.set_index(['mapel', 'test'])
df03
```
