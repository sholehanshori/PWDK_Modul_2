```python
# Membaca file dengan beberapa elemen isi yg kosong atau di isi simbol lain
df = pd.read_csv('dataCoba.csv')
df
```

```python
# Untuk mencari tau isi cell apakah kosong
df.isnull()    # Dapat ditulis   "df.isnull().sum()"
```

```python
# Untuk merespon data yg kosong tetapi terisi dengan (-) dan (n.a), maka dapat ditulis menjadi
df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
df
```

```python
# Mengisi cell kosong dengan nilai default yg telah ditentukan
df = df.fillna({
    'nama': 'Anonim',
    'usia': 24,
    'gaji': 5000000
})
df
```

```python
# Metode 'forwardfilling', yaitu mengisi cell yg kosong menggunakan konten cell yg sebelumnya

df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
# Forward Filling
df = df.fillna(method = 'ffill', axis='index')
df
```

```python
# Selain 'Forward Filling', ada juga 'Backward Filling'
df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
# Backward Filling
df = df.fillna(method = 'bfill', axis='index')
df
```

```python
# Dapat juga 'filling' secara horizontal
df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
# Forward Filling
df = df.fillna(method = 'ffill', axis=1)
df
```

```python
# Memastikan tipe dari data di dalam kolom

df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
df = df.fillna(0)

df['nama'] = df['nama'].astype(str)
df['gaji'] = df['gaji'].astype('int64')
df.dtypes
```

```python
# Penggunaan 'interpolate', untuk mengisi data yg kosong berdasarkan rata-rata data sebelum dan sesudahnya
# Hanya untuk data integer

df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
# Dalam 'interpolate' ditambahkan 'fillna' agar kolom 'nama' ada isi karena berupa 'string'
df = df.interpolate().fillna({
    'nama':'Anonim'
})
df
```

```python
# Apabila meng-drop data yg kurang/kosong, misal dikarenakan semua datanya diperlukan
df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
df = df.dropna()
df
```

```python
# Filter berupa jumlah isi minimal di dalam rows
df = df.dropna(thresh = 3)
df
```

```python
# Memfilter data, agar pada kolom 'nama' dan 'usia' tidak boleh kosong
df = pd.read_csv(
    'dataCoba.csv',
    na_values = ['-', 'n.a']
)
df = df.dropna(subset = ['nama', 'usia'])
df
```
