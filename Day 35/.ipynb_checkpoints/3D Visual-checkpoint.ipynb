```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
fig = plt.figure()
p   = plt.subplot(111, projection='3d')

x  = np.arange(10)
y  = np.arange(10)
z  = np.arange(10)

# Untuk visualisasi berupa garis
p.scatter(x, y, z)
plt.show()
```

```python
# Membentuk grafik bar 3 dimensi
fig = plt.figure()
p   = plt.subplot(111, projection='3d')

# Merepresentasikan titik mulai tiap sumbu axis
x  = np.arange(10)
y  = np.arange(10)
z  = np.zeros(10)

# Merepresentasikan value dari tiap sumbu axis
dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x, y, z, dx, dy, dz)
p.set_xlabel('Sumbu X')
p.set_ylabel('Sumbu Y')
p.set_zlabel('Sumbu Z')

plt.show()
```

```python
# Membentuk grafik bar 3 dimensi
fig = plt.figure(figsize=(12,8))     # figsize, merubah ukuran tampilan
p   = plt.subplot(111, projection='3d')

# Merepresentasikan titik mulai tiap sumbu axis
x  = np.ones(10)
y  = np.arange(10)
z  = np.zeros(10)

# Merepresentasikan value dari tiap sumbu axis
dx = np.ones(10)
dy = np.ones(10)
dz = np.arange(10)

p.bar3d(x, y, z, dx, dy, dz,
       color=['lightgreen']
)
p.set_xlabel('Sumbu X')
p.set_ylabel('Sumbu Y')
p.set_zlabel('Sumbu Z')

plt.show()
```

# Seaborn as sb

```python
# Menggunakan 'Seaborn'
# conda install -c anaconda seaborn
import seaborn as sb
df = pd.DataFrame({
    'x' : np.arange(10),
    'y' : np.arange(10),
    'z' : ['Pria', 'Wanita'] * 5
})

# Menggunakan style seperti 'seaborn'
plt.style.use('seaborn')
plt.plot(df.x, df.y)
print(df)
```

```python
# Memiliki bentuk yg sama dengan matplotlib
sb.relplot(data=df,  x='x', y='y', 
           hue='z', style='z', size='z'
          )
```

```python
# Menegggunakan 'seaborn' untuk grafik bar
sb.barplot(data=df, x='x', y='y')
plt.show()
```

```python
sb.barplot(data=df, x='x', y='y', hue='z')
plt.show()
```

```python
# Menggabungkan '.relplot' dan '.barplot'
sb.relplot(data=df,  x='x', y='y', 
           hue='z', style='z', size='z'
          )
sb.barplot(data=df, x='x', y='y')
```
