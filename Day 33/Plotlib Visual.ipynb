```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mengganti tampilan tema, cek untuk melihat pilihan tema ('plt.style.available')
plt.style.use('ggplot')

df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

# Penulisan data menggunakan 'key' dari 'set'
plt.plot(df['x'], df['y'], df.x, df.z)
plt.show()
```

```python
# Plot dapat juga ditulis seperti berikut
plt.plot(df['x'], df[['y', 'z']])

plt.grid(True)
# Penggunaan "ticks", untuk interval, custom value, atau rotasi pada sumbu x dan y
plt.xticks(np.arange(0, 7, step=1), list('abcdef'), rotation=60)
plt.yticks(np.arange(0, 10, step=1), rotation=90)
plt.show()
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

# Membuat chart berupa 'bar'
plt.figure('Tes')
plt.bar(df.x, df.z, color='black', edgecolor='gold', linewidth=2)
plt.xlabel('Ini X', fontsize=15, rotation=45)
plt.ylabel('Ini Y', fontsize=15, color='gold')

# Meng-edit title, berupa font style, size, atau color
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size': 30, 'color':'gold'
})

plt.show()
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

# Membuat chart berupa 'bar'
plt.bar(df.x, df.y, color='black', edgecolor='gold', linewidth=2)

# Membuat looping agar menampilkan value di atas setiap bar
for i in df['y']:
    plt.text(
        df[df['y']==i]['x'] -0.1, i+0.1, i,
        fontdict={'size':20}
    )

# Meng-edit title, berupa font style, size, atau color
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size': 30, 'color':'gold'
})

plt.show()
```

```python
# Stack 2 chart 'bar'
plt.bar(df.x, df.z, 
        color='black', 
        edgecolor='gold', linewidth=2)

plt.bar(df.x, df.y, 
        color='red', bottom= df.z,
        edgecolor='gold', linewidth=2)

# Membuat looping agar menampilkan value di atas setiap bar
for i in df['y']:
    plt.text(
        df[df['y']==i]['x'] -0.1, i+0.1, i,
        fontdict={'size':20}
    )

# Meng-edit title, berupa font style, size, atau color
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size': 30, 'color':'gold'
})

plt.show()
```

```python
# Memasukkan standar deviasi (STD) di tengah 'bar'
plt.bar(df.x, df.y, 
        color='grey',  yerr = df.describe()['y'].loc['std'],
         width=1, edgecolor='gold', linewidth=2)

# Membuat looping agar menampilkan value di atas setiap bar
for i in df['y']:
    plt.text(
        df[df['y']==i]['x'] -0.1, i+0.1, i,
        fontdict={'size':20}
    )

# Meng-edit title, berupa font style, size, atau color
plt.title('Ini bar chart', fontdict={
    'family':'impact', 'size': 30, 'color':'gold'
})

plt.show()
```

```python
# Membuat chart berupa 'pie'
presentase = [15, 40, 32, 13]
keterangan = ['Pak A', 'Pak B', 'Bu C', 'Om D']
plt.pie(
    presentase, labels=keterangan,
    colors=['gold', 'purple', 'salmon', 'orange'], startangle=0, shadow=True,
    counterclock=False, radius=1,
    textprops={'family':'dejavu sans','size': 11}, autopct='%1.1f%%'
)
plt.show()
```

```python
# Membuat chart berupa 'pie'
presentase = [15, 40, 32, 13]
keterangan = ['Pak A', 'Pak B', 'Bu C', 'Om D']

# Memberi variabel (x,y,z) untuk looping text prosentase, berdasarkan 'data raw'
# Menggunakan 'explode', untuk membuat "slice out" dari pie
x,y,z = plt.pie(
    presentase, labels=keterangan,
    colors=['gold', 'purple', 'salmon', 'orange'], startangle=180, shadow=True,
    counterclock=False, radius=1.5, explode=(.2,0,0,0),
    textprops={'family':'dejavu sans','size': 14}, autopct='%1.1f%%'
)

for i in z:
    i.set_color('w')
plt.show()
```

```python
# Membuat Histogram
usia  = [22,21,31,42,32,45,65,55,43,22,21,34,43,44]
usia2 = [33,31,41,21,22,25,25,55,54,52,41,44,23,24]
par   = [15,20,25,30,35,40,45,50,55,60,65]

plt.hist(
    [usia, usia2], par, color=['grey', 'black'], 
    rwidth=1, histtype='stepfilled' #backstacked
)
plt.xticks(np.arange(15, 65, step=5))
plt.show()
```
