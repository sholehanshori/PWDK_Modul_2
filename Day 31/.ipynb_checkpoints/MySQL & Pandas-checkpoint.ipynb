# Mengambil Data Sheet dari Telkom

```python
# Dari website https://finance.yahoo.com/quote/TLKM.JK/history?p=TLKM.JK
dfTlkm = pd.read_csv('TLKM.JK.csv')
dfTlkm
```

```python
# dfTlkm['Date'] = dfTlkm['Date'].astype('datetime64')
# Cek pada kolom 'Date' apakah berupa tipe waktu atau tidak
dfTlkm.dtypes
```

```python
# Dikarenakan tipe data dari kolom 'Date' adalah object, maka dilakukan
dfTlkm = pd.read_csv(
    'TLKM.JK.csv', parse_dates = ['Date'],
    index_col = 'Date'
)
dfTlkm
```

```python
dfTlkm.loc['2019-09-10']
```

```python
dfTlkm['2019-08-01':'2019-08-08']
```

```python
# Untuk mendapatkan data pada bulan tertentu
dfTlkm['2019-08']['Close']
```

```python
# Untuk mendapatkan detail data pada bulan tertentu, pada nilai max
dfTlkm[
    dfTlkm['Close'] ==
    dfTlkm['2019-08']['Close'].max()
]
```

```python
# Mendapatkan nilai rata-rata data
dfTlkm['2019-09']['Close'].resample('M').mean()
# Dengan M = Month End
# Dengan W = Week End
# Dengan Q = Quarter Year
# Dengan Y = Year
```

```python
# Merapikan data yg kosong.
dfTlkm = dfTlkm.resample('D').sum()
dfTlkm
```

```python
# Berdasarkan data, pada "Volume" awal tgl 5 - 7 kosong, dan ingin supaya terpisah maka
# Digunakan iloc untuk index 'nama kolom' [Date - Adj Close]

dfTlkm.iloc[:, :5] = dfTlkm.iloc[:, :5].replace(0, np.NaN) # Data yg kosong dirubah menjadi NaN
dfTlkm = dfTlkm.fillna(method='ffill')
dfTlkm
```
