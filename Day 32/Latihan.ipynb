```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dExcl = pd.read_csv('EXCL.JK.csv', parse_dates=['Date'])
dExcl = dExcl.fillna(method='ffill')

plt.figure(figsize=(12,7))
plt.plot(dExcl['Date'], dExcl['Close'], label='PT XL Axiata Tbk.' )

# plt.title('Testing')
plt.xlabel('Tanggal')
plt.ylabel('Rupiah (IDR)')
# plt.axis(ymin=0)
plt.xticks(rotation=45)
plt.suptitle('Harga Historis Saham Provider Telco Indonesia',fontsize=18, y=1.02)
plt.grid(True)
plt.legend(bbox_to_anchor=(0,1.01,1,0.78),loc='lower center',ncol=4,mode='expand')

plt.show()
```
